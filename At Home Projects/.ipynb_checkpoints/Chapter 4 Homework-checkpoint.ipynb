```python
pd.read_csv("2022-SP-Standard-Data-copy.csv")
df = pd.read_csv("2022-SP-Standard-Data-copy.csv")
df
```

```python
df["% of GS"] = df["GS"].div(df["G"]).mul(100)
df
```

```python
for column in df:
    print(column)
```

```python
def plot_df(df, age, IP):
    return plot_df
fig, ax = plt.subplots(figsize = (15,7))
ax.scatter(age, IP,
          color = ["C7"],
          alpha = .75,
          )
ax.set_title("Starting Pitcher IP vs Age Comparison")
ax.legend(["Individual SP"])
y1_ticklabels = [str(y) + " IP" for y in ax.get_yticks()]
ax.set_yticklabels(y1_ticklabels)
x_ticklabels = [str(x) + " Yrs" for x in ax.get_xticks()]
ax.set_xticklabels(x_ticklabels)
```
