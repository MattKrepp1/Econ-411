```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

oecd_countries = ["Australia", "Austria", "Belgium", "Canada", "Chile", "Colombia", "Costa Rica",
            "Czech Republic", "Denmark", "Estonia", "Finland", "France", "Germany",
            "Greece", "Hungary", "Iceland", "Ireland", "Israel", "Italy", "Japan", 
            "Latvia", "Lithuania","Luxembourg", "Mexico", "Netherlands", "New Zealand", "Norway",
             "Poland","Portugal", "Slovakia", "Slovenia", "South Korea", "Spain", "Sweden",
             "Switzerland", "Turkey", "United Kingdom", "United States"]

hw_data = pd.read_excel("Copy of EFWAndRGDP.xls")
hw_data[hw_data["Countries"].isin(oecd_countries)]
```

```python
iso = pd.read_excel("ISO3.xls")
iso
```

```python
del iso["Year"]
del iso["ISO_Code_2"]
del iso["World Bank Region"]
del iso["World Bank Current Income Classification, 1990-present (L=Low income, LM=Lower middle income, UM=Upper middle income, H=High income)"]
del iso["Summary"]
del iso["Size of Government"]
del iso["Legal System and Property Rights"]
del iso["Sound Money"]
del iso["Freedom to Trade Internationally"]
del iso["Regulation"]
del iso["RGDP Per Capita"]
iso
```

```python
iso = iso[iso["Countries"].isin(oecd_countries)]
iso
```

```python
iso = iso.drop_duplicates()
iso
```

```python
result = pd.merge(hw_data, iso, on="Countries")
result
```

```python
del result["ISO_Code_3_x"]
del result["World Bank Current Income Classification, 1990-present (L=Low income, LM=Lower middle income, UM=Upper middle income, H=High income)"]
```

```python
result = result.dropna()
```

```python
years_diff = 5
x_vars = result[['Size of Government', 'Legal System and Property Rights', 'Sound Money', 'Freedom to Trade Internationally', 'Regulation']]
y_var = result[['RGDP Per Capita']]
```

```python
r_df = result.dropna(axis = 0, how = "any")
y = y_var
x = x_vars
x["Constant"] = 1
results = sm.OLS(y, x).fit()
predictor = results.predict()
residuals = results.resid
results.summary()
```

```python
result["Residuals"] = results.resid
fig, ax = plt.subplots(figsize = (12,8))
result[["Residuals"]].plot.hist(bins = 100, ax = ax)
```

This histogram is different from the others because it shows that the residuals don't follow a normal distribution.

```python
results_dict = {"Beta":results.params,
               "t-stats": results.tvalues,
               "p-values":results.pvalues,
               "SE":results.bse}
results_df = pd.DataFrame(results_dict).round(3)
results_df.to_csv("y = RGPDPC, X = EFW, LogDiffResults.csv")
results_df
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(result['Size of Government'], predictor)
ax.set_xlabel('Size of Government')
ax.set_ylabel('Predictor')

fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(result['Legal System and Property Rights'], predictor)
ax.set_xlabel('Legal System and Property Rights')
ax.set_ylabel('Predictor')

fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(result['Sound Money'], predictor)
ax.set_xlabel('Sound Money')
ax.set_ylabel('Predictor')

fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(result['Freedom to Trade Internationally'], predictor)
ax.set_xlabel('Freedom to Trade Internationally')
ax.set_ylabel('Predictor')

fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(result['Regulation'], predictor)
ax.set_xlabel('Regulation')
ax.set_ylabel('Predictor')

plt.show()    
```
